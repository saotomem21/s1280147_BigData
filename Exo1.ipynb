```python
def extract_lat_lon(point_str):
    lon, lat = point_str.replace("Point(", "").replace(")", "").split()
    return float(lat), float(lon)

data['latitude'], data['longitude'] = zip(*data['Location'].map(extract_lat_lon))

import plotly.express as px

fig = px.scatter_mapbox(data, lat='latitude', lon='longitude', zoom=5, mapbox_style="open-street-map")
fig.show()
```
